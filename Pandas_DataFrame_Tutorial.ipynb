# Melting

```
data=pd.DataFrame({'Name':['A','B','C','D','E'], 'Age':[10,20,30,40,50],'Edu':['HS','MS',"",'HS',""]})
data
```

We are going to modify the abovedata with syntax learnt earlier.
1. break age into 3 group, under 25, between 25-50 and above <-- ***method using here is to create a new column, need to find how to replace directly in the "age" column without getting the error of comparing str vs. int***
2. remove na values and replace w/ previous values
3. concat this df with a new records
4. add a new income column and sort the data based on that colum


```
data_testing=data.copy()
data_testing.loc[data_testing['Age']<=25,'Group']='Under'
data_testing.loc[(data_testing['Age']<50) & (data_testing['Age']>25),'Group']='Middle'
data_testing.loc[data_testing['Age']>=50,'Group']='Senior'
data_testing
```

```
data_testing.loc[data_testing['Edu']=="",'Edu']=None
data_testing.isna()
```

```
#data_testing.dropna(axis=1)
data_testing.fillna(method='ffill',inplace=True)
data_testing
```

```
data.loc[data['Edu']=="",'Edu']=None
data['Group']='Vary'
combine_axis_0=pd.concat([data,data_testing],axis=0)
combine_axis_0
```

```
combine_axis_1=pd.concat([data,data_testing],axis=1)
combine_axis_1
```

Notice how we have to sort by a unique column, thus combine_axis_0 can short by Age but not combine_axis_1

```
combine_axis_0.sort_values(by='Age',ascending=True,inplace=True)
combine_axis_0
```

Pandas.melt() unpivots a DataFrame from wide format to long format.
melt() function is useful to massage a DataFrame into a format where one or more columns are identifier variables, while all other columns, considered measured variables, are unpivoted to the row axis, leaving just two non-identifier columns, variable and value.

pandas.melt(frame, id_vars=None, value_vars=None,
 var_name=None, value_name='value', col_level=None) ***bold text***

```
melt_df=pd.DataFrame({'Name': {0: 'John', 1: 'Bob', 2: 'Shiela'},
                   'Course': {0: 'Masters', 1: 'Graduate', 2: 'Graduate'},
                   'Age': {0: 27, 1: 23, 2: 21},
                      'ID':{0:1234,1:3456,2:5678}})
melt_df
```

```
pd.melt(melt_df,id_vars=['Name','ID'],value_vars=['Course'])
```

```
pd.melt(melt_df,id_vars=['ID'],value_vars=['Course','Age','Name'],var_name='Attributes',value_name='Stat')
```

Another way of making wider dataframe into narrower is stack and unstack()

```
stack_df=melt_df.stack()
stack_df
```

```
unstack_df=stack_df.unstack()
unstack_df
```

***pandas.pivot(index, columns, values) ***function produces pivot table based on 3 columns of the DataFrame. Uses unique values from index / columns and fills with values.

Parameters:
index[ndarray] : Labels to use to make new frames index
columns[ndarray] : Labels to use to make new frames columns
values[ndarray] : Values to use for populating new frames values

Returns: Reshaped DataFrame
Exception: ValueError raised if there are any duplicates.

```
melt_df
new_records=Series({'Name':'Tracy','Course':'HS','Age':18,'ID':'1000'},name='3')
melt_df=melt_df.append(new_records,ignore_index=False)
melt_df
```

```
melt_df.pivot(index=['ID'],values=['Age','ID'],columns=['Course'])
```

So we pick a new row index: ID
columns[]: how many distinct values in this " Course" will be the new columns
values[]: based on the ID, the columns, which values to fill in the slot.

```
melt_df # modify melt_df to have a duplicated ID then perform pivot to see the missing index
melt_df=melt_df.append(Series({'Name':'John','Course':'HS','Age':19,'ID':'1234'},name='4'))
melt_df
```

Now let perform the pivot again with the duplicated value in the "ID" col, which will be used as index for the new table

```
dup_index_pivot=melt_df.pivot(index=['ID','Name'],columns=['Course'],values=['Age'])
# noticing how the ID for second John is missing
```

```
dup_index_pivot.reset_index()
# reset index release the missing ID since it is no longer the row index of the table
```
